```python
def get_html(x):
    try:
        ret = re.findall('(http://.*)',x)[0]
#         ret=
    except:
        ret = None
    return ret
```

```python
df['url'] = df['qr_res'].map(get_html)
```

```python
df.to_excel('new_rsult.xls')
```

```python
def get_host(x):
    try:
        ret = re.findall('http://(.*?)/',x)[0]
    except:
        ret =None
    return ret
```

```python
df['host']=df['url'].map(get_host)
```
