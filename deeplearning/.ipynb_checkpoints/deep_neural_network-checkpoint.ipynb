```python
import numpy as np
import neurolab as nl
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
data = np.array([
    [0.3,0.2],
    [0.1,0.4],
    [0.4,0.6],
    [0.9,0.5]
])
```

```python
labels=np.array([
    [0],
    [0],
    [0],
    [1]
])
```

```python
plt.figure()
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
perceptron = nl.net.newp([[0,1],[0,1]],1)
```

```python
error = perceptron.train(data,labels,epochs=50,show=15,lr=0.01)
```

```python
plt.figure()
plt.plot(error)
plt.xlabel('number of epochs')
plt.ylabel('training error')
plt.grid()
plt.title('network')
plt.show()
```

```python
# 多层神经网络
min_value =-12
max_value =12
num_datapoints=90
```

```python
x= np.linspace(min_value,max_value,num_datapoints)
```

```python
x
```

```python
len(x)
```

```python
y=2*np.square(x)+7
```

```python
y/=np.linalg.norm(y)
```

```python
plt.figure()
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
data=x.reshape(num_datapoints,1)
labels=y.reshape(num_datapoints,1)
```

```python
plt.figure()
plt.scatter(data[:,0],labels[:,0])
plt.title('scatter')
plt.grid()
plt.show()
```

```python
data
```

```python
multi_layer = nl.net.newff([[min_value,max_value]],[10,10,1])
```

```python
multi_layer.trainf=nl.train.train_gd
```

```python
error = multi_layer.train(data,labels,epochs=800,show=100,goal=0.01)
```

```python
plt.figure()
plt.plot(error)
plt.title('error')
plt.grid()
plt.show()
```

```python
x2=np.linspace(min_value,max_value,num_datapoints*2)
y2=multi_layer.sim(x2.reshape(num_datapoints*2,1)).reshape(x2.size)
```

```python
plt.figure()
plt.plot(x2,y2,'+',x,y,'.')
```
