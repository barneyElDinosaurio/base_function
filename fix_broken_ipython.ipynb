```python
# 拯救损坏的jupyter 文件
import re
import codecs
```

```python
pattern = re.compile('"source": \[(.*?)\]\s+\},',re.S)
```

```python
filename = 'tushare_sample.ipynb'
with codecs.open(filename,encoding='utf8') as f:
    content = f.read()
```

```python
content
```

```python
source = pattern.findall(content)
for s in source:
    t=s.replace('\\n','')
    t=re.sub('"','',t)
    t=re.sub('(,$)','',t)
    print(t)
```
