```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import pytesseract
```

```python
from PIL import Image
```

```python
im = Image.open('tw_photo.jpg')
im = im.convert('L')
plt.imshow(im)
```

```python
im=np.array(im)
ret,new_img = cv2.threshold(im,127,255,0)
```

```python
plt.imshow(new_img)
```

```python
pytesseract.image_to_string(new_img)
```
