```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
from PIL import Image
```

```python
file = '2999.gif'
try:
    img = cv2.imread(file)
except Exception as e:
    print(e)
# plt.imshow(img)
```

```python
kernel = np.ones((1,1),np.uint8)
erosion = cv2.erode(img,kernel,iterations =1 )
plt.imshow(erosion)
```

```python
# read gif
im = Image.open('2999.gif')
# im.show()
im=im.convert('RGB')
```

```python
plt.imshow(im)
# img.shape
```

```python
im_array = np.asarray(im)
```

```python
im_array
```

```python
cv_img = cv2.cvtColor(im_array,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(cv_img)
```

```python
kernel = np.ones((1,1),np.uint8)
erosion = cv2.erode(cv_img,kernel,iterations =4 )
plt.imshow(erosion)
```

```python
resize = 4
img_scale = cv2.resize(cv_img,None,fx=resize,fy=resize,interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(img_scale)
```

```python
kernel = np.ones((1,1),np.uint8)
erosion = cv2.erode(cv_img,kernel,iterations =1 )
plt.imshow(erosion)
```

```python
canny = cv2.Canny(cv_img,50,40)
plt.imshow(canny)
```
