```python
# 
import numpy as np
```

```python
img_lg = cv2.imread(os.path.join(IMAGE_FOLDER,'box.png'))
```

```python
plt.imshow(img_lg)
plt.title('origin')
```

```python
img_lg_gray = cv2.cvtColor(img_lg,cv2.COLOR_BGR2GRAY)
```

```python
img_lg_gray_f = np.float32(img_lg_gray)
```

```python
img_haris = cv2.cornerHarris(img_lg_gray_f,7,5,0.04)
```

```python
plt.imshow(img_haris)
```

```python
img_haris=cv2.dilate(img_haris,None)
```

```python
img_lg[img_haris>0.01*img_haris.max()]=[0,0,0]
```

```python
plt.imshow(img_lg)
plt.title('corner')
```
