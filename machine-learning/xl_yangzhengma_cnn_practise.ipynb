```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
X_train_4D = X_train.reshape(X_train.shape[0],12,15,1).astype('float32')
X_test_4D = X_test.reshape(X_test.shape[0],12,15,1).astype('float32')
```

```python
X_train_4D_normal=X_train_4D/255
X_test_4D_normal=X_test_4D/255
```

```python
from keras.utils import np_utils
```

```python
y_train_onehot = np_utils.to_categorical(y_train)
y_test_onehot = np_utils.to_categorical(y_test)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
```

```python
model = Sequential()
```

```python
model.add(
   Conv2D(
   filters=16,
   kernel_size=(5,5),
   padding='same',
   input_shape=(12,15,1),
   activation='relu',
   )
)
```

```python
model.add(
MaxPooling2D(pool_size=(2,2))
)
```

```python
model.add(
Conv2D(
filters=36,
    kernel_size=(5,5),
    padding='same',
    activation='relu',
)
)
```

```python
model.add(
MaxPooling2D(pool_size=(2,2))
)
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
```

```python
model.add(
Dense(
128,activation='relu'
)
)
```

```python
model.add(Dropout(0.5))
```

```python
model.add(
Dense(
10,
    activation='softmax'
)
)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history_train = model.fit(x=X_train_4D_normal,
                         y=y_train_onehot,
                         validation_split=0.2,
                          epochs=10,
                          batch_size=300,
                          verbose = 2,
                         )
```

```python
score = model.evaluate(X_test_4D_normal,y_test_onehot)
```
