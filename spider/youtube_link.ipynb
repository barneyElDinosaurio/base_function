```python
import codecs
with codecs.open('youtube.txt','r',encoding='utf8') as f:
    content = f.read()
```

```python
import re
```

```python
ret = re.findall('\{\"url\":\"/watch\?v=(\S+?)",',content)
url='https://www.youtube.com/watch?v='
result = []
for i in ret:
    print(i)
#     print()
    result.append(url+i.strip())
```
